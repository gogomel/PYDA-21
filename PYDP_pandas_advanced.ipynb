```python
"""Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

1. если источник traffic_source равен yandex или google, то в source_type ставится organic
2. для источников paid и email из России - ставим ad
3. для источников paid и email не из России - ставим other
4, все остальные варианты берем из traffic_source без изменений"""
```

```python
log = pd.read_csv('visit_log.csv', sep=';')

log.loc[(log.traffic_source  == 'yandex')| (log.traffic_source  == 'googel'), 'source_type'] = 'organic' # 1 пункт задания
log.loc[(log.region == 'Russia')&(log.traffic_source  == 'paid')|(log.traffic_source  == 'email'), 'source_type'] = 'ad' # 2 пункт задания
log.loc[(log.region != 'Russia')&(log.traffic_source  == 'paid')|(log.traffic_source  == 'email'), 'source_type'] = 'other' # 3 пункт задания
log.source_type = log.source_type.fillna(log.traffic_source) # 4 пункт задания

log.head(10)
```

```python
"""Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница между максимальным и минимальным значением 
столбца timestamp для данного значения userId."""
```

```python
ratings = pd.read_csv('ratings.csv')

ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s') # добавляю колонку даты 
new_ratings = ratings.groupby('userId').count().query('rating > 100').reset_index() # делаю датафрейм, где будет пользователи, которые поставили больше 100 оценок
dict_ = {} # словарь для ответа (ключи - пользователи, значения - среднее время жизни)
for row in new_ratings['userId']:  
    time = ratings[ratings.userId == row]['date'].max() - ratings[ratings.userId == row]['date'].min() # здесь я по  отфильтрованному датафрейму высчитываю время жизни в начальном датафрейму
    dict_[row] = time
#a = pd.DataFrame([dict_]) 
a = pd.DataFrame.from_dict(dict_, orient='index').rename(columns={0:'date'})# преобразую в человекочитаемый вид как могу
new_ratings.head()
ratings.head()
```

```python
"""Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в 
разделе Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).

Необходимо сформировать две таблицы:
1. таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
2. аналогичную таблицу по типам выручки с указанием адреса клиента"""
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
task_1 = rzd.merge(auto, how = 'outer', on = 'client_id').merge(air, how = 'outer', on = 'client_id') # пернвый пункт
task_1
```

```python
task_2 = task_1.merge(client_base, how = 'left', on = 'client_id') # второй пункт
task_2
```
