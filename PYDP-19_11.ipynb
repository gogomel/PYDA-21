```python
"""Задание 2
Используем файл keywords.csv.
Необходимо написать гео-классификатор, который каждой строке сможет
выставить географическую принадлежность определенному региону. 
Т. е. если поисковый запрос содержит название города региона, 
то в столбце ‘region’ пишется название этого региона. Если поисковый 
запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
           }
"""
```

```python
df_2 = pd.read_csv('keywords.csv')
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
           }
```

```python
def region_name(dataframe):
    for region, city in geo_data.items():
        for el in city:
            if el in dataframe['keyword']:
                return region
    return 'undefined'
```

```python
df_2['region'] = df_2.apply(region_name, axis=1)
df_2[df_2.region == 'Дальний Восток'].head()
```
